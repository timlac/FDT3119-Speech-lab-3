### 4.3 Feature Extraction

```python
phoneHMMs = np.load('lab2_models_all.npz', allow_pickle=True)['phoneHMMs'].item()
phones = sorted(phoneHMMs.keys())
nstates = {phone: phoneHMMs[phone]['means'].shape[0] for phone in phones}
```

```python
with open('stateList.pkl', 'rb') as file:
    stateList = pickle.load(file)
```

```python
traindata = []

for root, dirs, files in os.walk(os.path.join(tidigits_path, 'disc_4.1.1/tidigits/train')):
    print(f"Processing directory: {root}")
    for file in files:
        if file.endswith('.wav'):
            filename = os.path.join(root, file)
            samples, samplingrate = loadAudio(filename)

            lmfcc = mfcc(samples)
            _mspec = mspec(samples)

            wordTrans = list(path2info(filename)[2])
            phoneTrans = words2phones(wordTrans, prondict, addSilence=True, addShortPause=True)

            target_states = forcedAlignment(lmfcc, phoneHMMs, phoneTrans, nstates)

            targets = []
            for state in target_states:
                targets.append(stateList.index(state))

            traindata.append({'filename': filename, 'lmfcc': lmfcc,
                              'mspec': _mspec, 'targets': targets})

np.savez('data/traindata.npz', traindata=traindata)
```
